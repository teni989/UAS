```python
from sklearn import tree
import pandas as pd
import pydotplus
from IPython.display import Image
```

```python
atribut = ['jenis_kelamin', 'agama', 'jurusan', 'asal_sekolah', 'kota_asal', 'registrasi']
ds = pd.read_csv('siswa-dataset.csv')
ds.head(15)
```

```python
siswa = pd.get_dummies(ds[['jenis_kelamin', 'agama', 'jurusan', 'asal_sekolah', 'kota_asal']])
siswa.head()
```

```python
hasil = tree.DecisionTreeClassifier(criterion='entropy')
hasil_train = hasil.fit(siswa, ds['registrasi'])
```

```python
dot_data = tree.export.graphviz(hasil_train, out_file-None, feature_names=list(siswa.columns.values),
                                class_names=['TIDAK', 'YA'], rounded=True, filled=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
